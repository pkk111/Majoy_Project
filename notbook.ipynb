```python
import tensorflow as tf

from tensorflow.keras import layers
from keras.preprocessing.image import ImageDataGenerator
from numpy import expand_dims
import cv2
import os
import math

def data_agument(image):
    datagen = ImageDataGenerator(featurewise_center=True,rotation_range=(0-30),width_shift_range=0.1,
                             height_shift_range=0.1,brightness_range=[0.75,1.25],
                             shear_range=0.2, channel_shift_range=0.2,
                             horizontal_flip=True, vertical_flip=True,fill_mode='nearest')

    it = datagen.flow(expand_dims(image, 0), batch_size=1)
    
    images = []
    for x in range(10):
      batch = it.next()
      img = batch[0].astype('uint8')
      images.append(img)
    return images

# record = [2, "C140P101ThinF_IMG_20151005_210207_cell_102.png"]
# path = base_dir + labels[record[0]] + '/' + record[1]
# image = cv2.imread(path)
# images = data_agument(image)
# for i in range(len(images)):
#     path = base_dir + 'augmented/' + labels[record[0]] + f"/{record[1].replace('.png', '')}"
#     result = cv2.imwrite(path + f"_{i}.png", images[i])
#     print(path + f"_{i}.png")


aug_images = []
for record in data:
  if (type(record[1]) == float and (math.isnan(record[1]) or record[1] in ['nan', np.nan, ... ])):
    print(type(record[0]), record[0], type(record), record[1])
    continue
  path = base_dir + labels[record[0]] + '/' + record[1]
  image = cv2.imread(path)
  images = data_agument(image)
  for i in range(len(images)):
    path = base_dir + 'augmented/' + labels[record[0]] + f"/{record[1].replace('.png', '')}"
    result = cv2.imwrite(path + f"_{i}.png", images[i])
```
